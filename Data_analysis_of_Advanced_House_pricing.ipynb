## Handling missing values

```python
#making list of features iwth nan values
features_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>0]
#displaying feature with nan and their percentage wrt all values
for features in features_with_nan:
  print(features,np.round(dataset[features].isnull().mean(),4),"% missing values")
```

###As there are many missing values we need to find relationship between missing values and sales price

```python
for feature in features_with_nan:
  data=dataset.copy()
  #lets make a variable that indicates 1 if the observations was missing or 0
  data[feature]=np.where(data[feature].isnull(),1,0)
# calculate the mean salesprice where the information is missing or present
  data.groupby(feature)['SalePrice'].median().plot.bar(color=["red","blue"])
  plt.title(feature)
  plt.show()
  
```

###observation:
 1.As there is a strong relation between missing values and dependant values . so we need to replace with something meaningful value

 2. from dataset we can drop "id" column 

##Numerical variables

```python
#list of numerial variables 
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes != 'object']
print("Number of numerical variables:",len(numerical_features))

#displaying numerical_variables
dataset[numerical_features].head()
```

###Temporal variables (datetime variable) extraction

```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or "Year" in feature]
print(year_feature)
```

```python
# display contents of year variable
for feature in year_feature:
  print(feature,dataset[feature].unique())
```

```python
## lets analyze the temporal datetime variable, i.e we will check whether there is a relationship bw year_sold and saleprice
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('year sold')
plt.ylabel('Median House price')
plt.show()
```

###observation :
1.As Yearsold increases respective sale price decreases

### we will analyse with all other year variables

```python
## we will compare the difference between all years and yrsold
for feature in year_feature:
  if feature!='YrSold':
    data=dataset.copy()
    data[feature]=data['YrSold']-data[feature]

    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel("Yearsold-"+feature)
    plt.ylabel('SalePrice')
    plt.show()
```

###observation:

1.As the difference increases then saleprice goes down

##Discrete feature

```python
##numerical varibles are usually of 2 types
##1.continous variable and Discrete variables

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and (feature not in (year_feature or ['Id']))]

print(discrete_feature)
print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
# finding relationship bw discrete_features and saleprice
for feature in discrete_feature:
  data=dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar(color=['red', 'yellow', 'black', 'blue', 'orange'])
  plt.xlabel(feature)
  plt.ylabel('SalePrice')
  plt.title(feature)
  plt.show()
```

```python
#continuous variable

continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+["Id"]]
print(len(continuous_feature))
```

```python
# lets analyse the continuous_feature 

for feature in continuous_feature:
  data=dataset.copy()
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel('count')
  plt.title(feature)
  plt.show()
```

##Logarithmics transformation

```python
for feature in continuous_feature:
  
    data[feature]=np.log1p(dataset[feature])     #applying logarithmic function
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()  
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(feature)
    plt.show()             
                         
```

##Handling outliers

```python
for feature in continuous_feature:
  data=dataset.copy()
  data[feature]=np.log1p(data[feature])
  data.boxplot(column=feature)
  plt.ylabel(feature)
  plt.title(feature)
  plt.show()  
```
